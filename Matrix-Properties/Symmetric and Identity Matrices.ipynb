# Symmetric Matrix

A symmetric matrix is a square matrix that is equal to its own transpose. This means the matrix remains unchanged when reflected across its main diagonal. Mathematically, for a matrix A to be symmetric, the element in row i and column j (aij) must be equal to the element in row j and column i (aji) for all i and j. 

[Symmetric and Skew Symmetric Matrices](https://www.geeksforgeeks.org/maths/what-is-symmetric-matrix-and-skew-symmetric-matrix/)

For example, let us consider a square matrix
