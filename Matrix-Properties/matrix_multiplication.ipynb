# Matrix Multiplication (with a Vector)

## Numpy

```python
import numpy as np
```

```python
A = np.array([[3,4],[5,6],[7,8]])
B = np.array ([1,2])
```

```python
np.dot(A, B)
```

```python
C = np.array([[3,4],[5,6],[7,8]])
D = np.array([[1,9],[2,0]])
```

```python
np.dot(C, D)
```

## Torch

```python
import torch
```

```python
A_pt = torch.tensor([[3,4],[5,6],[7,8]])
B_pt = torch.tensor([1,2])
```

```python
torch.matmul(A_pt,B_pt)
```

```python
C_pt = torch.tensor([[3,4],[5,6],[7,8]])
D_pt = torch.tensor([[1,9],[2,0]])
```

```python
torch.matmul(C_pt, D_pt)
```

## Tensorflow

```python
import tensorflow as tf
```

```python
A_tf = tf.Variable([[3,4],[5,6],[7,8]])
B_tf = tf.Variable([1,2])
```

```python
tf.linalg.matvec(A_tf, B_tf)
```

```python
C_tf = tf.Variable([[3,4],[5,6],[7,8]])
D_tf = tf.Variable([[1,9],[2,0]])
```

```python
tf.linalg.matmul(C_tf, D_tf)
```
