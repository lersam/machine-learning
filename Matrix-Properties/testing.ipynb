# Eigenvectors and Eigenvalues

```python
A = np.array([[-1,  4], [ 2, -2]])
lambdas, V = np.linalg.eig(A)
print("lambdas: ",lambdas)
print("V: ", V)
```

```python
np.dot(A,V)
```

```python
np.matmul(lambdas, V)
```

```python
np.dot(A,V[:,0])
```

```python
np.dot(A,V[:,1])
```

```python
np.dot(A,V[:,0]) == np.dot(lambdas[0], V[:,0])
```

```python
np.dot(A,V[:,1]) == np.dot(lambdas[1], V[:,1])
```

# Copilot examples

To calculate the **determinant of a 3Ã—3 matrix**, we use a method called **cofactor expansion**. Here's how to do it step by step for your matrix:
$$
A = \begin{bmatrix}
2 & 1 & -3 \\
4 & -5 & 2 \\
0 & -1 & 3
\end{bmatrix}
$$
---

### ðŸ§® Step-by-Step Determinant Calculation

Weâ€™ll expand along the **first row** (you can choose any row or column, but the first row is often easiest):

$$
\text{det}(A) = a_{11} \cdot C_{11} - a_{12} \cdot C_{12} + a_{13} \cdot C_{13}
$$

Where:
- \( a_{ij} \) is the element in row *i*, column *j*
- \( C_{ij} \) is the determinant of the 2Ã—2 minor matrix formed by removing row *i* and column *j*

---

#### ðŸ”¹ Step 1: Expand the determinant

$$
\text{det}(A) = 2 \cdot \begin{vmatrix} -5 & 2 \\ -1 & 3 \end{vmatrix}
- 1 \cdot \begin{vmatrix} 4 & 2 \\ 0 & 3 \end{vmatrix}
+ (-3) \cdot \begin{vmatrix} 4 & -5 \\ 0 & -1 \end{vmatrix}
$$

---

#### ðŸ”¹ Step 2: Calculate each 2Ã—2 determinant
$$
- \begin{vmatrix} -5 & 2 \\ -1 & 3 \end{vmatrix} = (-5)(3) - (2)(-1) = -15 + 2 = -13
$$
$$
- \begin{vmatrix} 4 & 2 \\ 0 & 3 \end{vmatrix} = (4)(3) - (2)(0) = 12 - 0 = 12
$$
$$
- \begin{vmatrix} 4 & -5 \\ 0 & -1 \end{vmatrix} = (4)(-1) - (-5)(0) = -4 - 0 = -4
$$
---

#### ðŸ”¹ Step 3: Plug values back in

$$
\text{det}(A) = 2(-13) - 1(12) + (-3)(-4)
= -26 - 12 + 12 = -26
$$

---

### âœ… Final Answer

$$
\boxed{\text{det}(A) = -26}
$$

Yes, Samuel! For a **3Ã—3 matrix**, there's a shortcut called the **diagonal method** or **Sarrus' Rule** â€” it's a neat trick that works only for 3Ã—3 matrices. Here's how it works for your matrix:

$$
A = \begin{bmatrix}
2 & 1 & -3 \\
4 & -5 & 2 \\
0 & -1 & 3
\end{bmatrix}
$$

---

### ðŸ”· Step-by-Step: Diagonal Method (Sarrus' Rule)

#### ðŸ”¹ Step 1: Write the matrix and repeat the first two columns

Extend the matrix by writing the first two columns again to the right:

$$
\begin{array}{ccc|cc}
2 & 1 & -3 & 2 & 1 \\
4 & -5 & 2 & 4 & -5 \\
0 & -1 & 3 & 0 & -1 \\
\end{array}
$$

---

#### ðŸ”¹ Step 2: Add the products of the **three downward diagonals**
$$
D_1 = 2 \cdot (-5) \cdot 3 = -30 \\
D_2 = 1 \cdot 2 \cdot 0 = 0 \\
D_3 = -3 \cdot 4 \cdot (-1) = 12
$$
Sum of downward diagonals:
$$
S_1 = -30 + 0 + 12 = -18
$$
---

#### ðŸ”¹ Step 3: Subtract the products of the **three upward diagonals**

$$
U_1 = 0 \cdot (-5) \cdot (-3) = 0 \\
U_2 = -1 \cdot 2 \cdot 2 = -4 \\
U_3 = 3 \cdot 4 \cdot 1 = 12
$$
Sum of upward diagonals:  
$$
S_2 = 0 + (-4) + 12 = 8
$$
---

#### ðŸ”¹ Step 4: Final Determinant
$$
\text{det}(A) = S_1 - S_2 = -18 - 8 = -26
$$
---

### âœ… Final Answer
$$
\boxed{\text{det}(A) = -26}
$$
Same result as the cofactor method â€” just faster for 3Ã—3! Want to try this method on another matrix or see how it breaks down visually?
