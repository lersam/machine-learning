Matrix inversion is the process of finding an inverse matrix, denoted A⁻¹, such that when it is multiplied by the original matrix A, the result is the identity matrix (I)

$X^{-1}X = I_n$



## Numpy

```python
import numpy as np
```

```python
X = np.array([[4, 2], [-5, -3]])
# Y = np.array([[1.5, 1],[-2.5, -2]])

X_inv = np.linalg.inv(X)
X_inv
```

```python
Y = np.array([4, -7])
Y
```

```python
w = np.dot(X_inv, Y)
w
```

```python
np.dot(X,w)
```

## Torch

```python
import torch
```

```python
X = torch.tensor([[4,  2], [-5, -3.]])
Y = torch.tensor([4, -7.])
```

```python
X_inv = torch.inverse(X)
X_inv
```

```python
Y
```

```python
w = torch.matmul(X_inv, Y)
w
```

## Tensorflow

```python
import tensorflow as tf
```

```python
X = tf.Variable([[4, 2], [-5, -3.]])
Y = tf.Variable([4, -7.])
```

```python
X_inv = tf.linalg.inv(X)
X_inv
```

```python
w = tf.linalg.matvec(X_inv, Y)
w
```
