```python
from sklearn.datasets import fetch_openml
import matplotlib.pyplot as plt

# Load data from www.openml.org
X, y = fetch_openml("mnist_784", version=1, return_X_y=True, as_frame=False)
```

```python
# Display a grid of images from X.
# Set `n` to the number of images to show (warning: large values can be slow).
n = 25  # try 25, or set to len(X) to show all (not recommended)
n_cols = 5
import math
n_rows = math.ceil(n / n_cols)
plt.figure(figsize=(n_cols * 2, n_rows * 2))
for i in range(min(n, X.shape[0])):
    ax = plt.subplot(n_rows, n_cols, i + 1)
    img = X[i].reshape(28, 28)
    plt.imshow(img, cmap='gray')
    plt.title(str(y[i]))
    plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
index = 250
# Display an image (optional)
# Reshape the 784-pixel flat array into a 28x28 image
image_data = X[index].reshape(28, 28)
plt.imshow(image_data, cmap='gray')
plt.title(f"Label: {y[index]}")
plt.show()

# X contains the image data (as a 2D array of samples x pixels)
# y contains the labels (the actual number for each image)
```
